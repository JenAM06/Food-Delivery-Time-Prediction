### Penanganan Missing Value

Penanganan missing values dilakukan dengan dua cara, yaitu imputasi dan dihapus, dengan ketentuan:

*   Dihapus jika data yang hilang atau tidak terbaca mencapai >= 70%
*   Dapat dihapus atau diimputasi apabila data yang hilang masih <=70%, diimputasi dengan cara mengisi data yang kosong menggunakan mean, median, modus dari kolom terkait

#### IMPUTASI MISSING VALUE

Melakukan Imputasi pada data *categorical* dengan metode modus untuk mengisi data yang hilang. Kolom yang akan diimputasi dengan metode modus adalah 'Weather', 'Traffic_Level' dan 'Time_of_Day'.

```python
df['Weather'] = df['Weather'].fillna(df['Weather'].mode()[0])
df['Traffic_Level'] = df['Traffic_Level'].fillna(df['Traffic_Level'].mode()[0])
df['Time_of_Day'] = df['Time_of_Day'].fillna(df['Time_of_Day'].mode()[0])
```

## OUTLIERS

Outliers adalah nilai yang jauh dari distribusi data utama. Penanganannya meliputi penghapusan, transformasi, atau clipping agar tidak memengaruhi analisis.

Memeriksa outliers

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Data "Delivery_Time_min" dengan persentase 0,6 menggunakan Winsorizing (clipping) karena metode ini menggantikan nilai ekstrem dengan batas tertentu (persentil 0,6 dalam kasus ini). Hal ini membantu menjaga distribusi data tetap representatif tanpa menghapus data, sehingga analisis tetap valid tanpa dipengaruhi outlier ekstrem.

```python
columns_to_impute = ["Delivery_Time_min"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    print(f"{col} â†’ Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")

    # Menggunakan .loc[] agar tidak muncul SettingWithCopyWarning
    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

Pengecekan kembali

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

# CONSTRUCT DATA

Construct data adalah proses membuat fitur baru dari data yang ada untuk meningkatkan analisis atau model.

## Mengonversi Data Kategorikal ke Numerik
Agar dapat digunakan dalam analisis statistik dan model prediksi, Data kategorikal ini perlu dikonversi menjadi numerik.

```python
# Kolom 'Weather'
df["Weather"] = df["Weather"].map({"Clear": 0, "Windy": 1, "Foggy": 2, "Rainy": 3, "Snowy": 4})

# Kolom 'Time_of_Day'
df["Time_of_Day"] = df["Time_of_Day"].map({"Morning": 0, "Afternoon": 1, "Evening": 2, "Night": 3})

# Kolom 'Traffic_Level'
df["Traffic_Level"] = df["Traffic_Level"].map({"Low": 0, "Medium": 1, "High": 2})

# Kolom 'Vehicle_Type'
df["Vehicle_Type"] = df["Vehicle_Type"].map({"Scooter": 0, "Bike": 1, "Car": 2})
```
